```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = '/Users/nelsschimek/Documents/codeProjects/basketball_analysis/wehoop-wnba-data/wnba/player_box/csv/'
```

```python
historical_data = pd.read_csv(data_path + 'all_player_scoring_data.csv')
historical_data.head()
```

sort(historical_data['player_names'].unique())

```python
top_scoring = historical_data[historical_data['player_ppg'] > 20]
top_scoring
```

```python
aja_data = historical_data[historical_data['player_names'] == 'A\'ja Wilson']
aja_data_2024 = aja_data[aja_data['player_seasons'] == 2024]
aja_data_2023 = aja_data[aja_data['player_seasons'] == 2023]
aja_data_2024
```

```python
dt_data = historical_data[historical_data['player_names'] == 'Diana Taurasi']
dt_data_best_scoring = dt_data[dt_data['player_ppg'] > 20]
dt_data_best_scoring
```

```python
plt.scatter(x=historical_data['player_ppg'], y=historical_data['player_ts'], s=4)
plt.scatter(x=aja_data_2024['player_ppg'], y=aja_data_2024['player_ts'], label='A\'ja Wilson 2024')
plt.scatter(x=aja_data_2023['player_ppg'], y=aja_data_2023['player_ts'], label='A\'ja Wilson 2023')
plt.scatter(x=dt_data_best_scoring['player_ppg'], y=dt_data_best_scoring['player_ts'],label = 'Diana Taurasi', s=10)
plt.xlabel('Points per game')
plt.ylabel('True Shooting %')
plt.legend()
```
